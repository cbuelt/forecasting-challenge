```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, datetime, timedelta
import random
from scipy import stats
import ephem
```

```python
quantiles = [0.025, 0.25, 0.5, 0.75, 0.975]
horizons = [36, 48 ,60, 72, 84]
sns.set_theme(color_codes=True)
```

```python
data = pd.read_feather("data/berlin_data/historic_data/icon_eps_t_2m.feather")
#Filter for horizons
data = data[data["fcst_hour"].isin(horizons)]
data.dropna(inplace=True)
data.head()
```

## Calculate Forecast error

```python
data["Forecast_error"] = data["obs"]-data["ens_mean"]
```

# Visualize relationship between time and forecast error

```python
fig, ax = plt.subplots(figsize=(12,10))
sns.lineplot(x = "obs_tm", y = "Forecast_error", hue = "fcst_hour", data=data)
```

```python
data["day"] = pd.DatetimeIndex(data["obs_tm"]).dayofyear
plot = data.groupby("day").mean().reset_index()
fig, ax = plt.subplots(figsize=(12,10))
sns.regplot(x = "day", y = "Forecast_error", data=plot)
print("Pearson-Correlation: \t {}\nSpearman_correlation: \t {}".format(stats.pearsonr(plot["day"], plot["Forecast_error"]),stats.spearmanr(plot["day"], plot["Forecast_error"])))
```

## Analysis
It is evident that there is some correlation between the forecast error and the time of the year. The correlation is small, but timely information could be implemented into the temperature model.

# Visualize relationship between Forecast error and wind speed forecast

```python
wind_data = pd.read_feather("data/berlin_data/historic_data/icon_eps_wind_10m.feather")
wind_data.dropna(inplace=True)
wind_data=wind_data[["obs_tm","ens_mean","ens_var","fcst_hour"]]
wind_data = wind_data[wind_data["fcst_hour"].isin(horizons)]
wind_data.rename(columns = {"ens_mean":"wind_mean", "ens_var":"wind_var"},inplace=True)
```

```python
data_merged = data.merge(wind_data, left_on = ["obs_tm","fcst_hour"], right_on = ["obs_tm","fcst_hour"], how = "inner")
```

```python
fig, ax = plt.subplots(2,3,figsize=(25,15))
ax = ax.ravel()
for cnt, horizon in enumerate(horizons):
    #Filter horizon
    data_plot = data_merged[data_merged["fcst_hour"]==horizon].copy()

    sns.regplot(x = "Forecast_error", y = "wind_mean", data=data_plot, ax = ax[cnt])
    ax[cnt].set_xlabel("Forecast error",size = 15)
    ax[cnt].set_ylabel("Wind mean estimation", size = 17)
    ax[cnt].set_title("Correlation for horizon = {}".format(horizon),size = 20)
    r = stats.pearsonr(data_plot["wind_mean"], data_plot["Forecast_error"])[0]
    rho = stats.spearmanr(data_plot["wind_mean"], data_plot["Forecast_error"])[0]
    ax[cnt].annotate('Correlation:\nr = {:.4f}\n'.format(r)+r'$\rho$ = {:.4f}'.format(rho),xy = (.01,.02), xycoords=ax[cnt].transAxes, size = 15)
```

# Visualize relationship with angle of the sun

```python
sun = ephem.Sun()
observer = ephem.city("Berlin")
```

```python
def get_angle(observer, sun, timestamp):
    date = timestamp[0].strftime("%Y/%m/%d %H:%M:%S")
    observer.date = date
    sun.compute(observer)
    return sun.az
```

```python
angle_df = pd.DataFrame(data["obs_tm"].unique(),columns = ["obs_tm"])
angle_df["degree"] = angle_df.apply(lambda x: get_angle(observer,sun,x), axis = 1)
```

```python
data_merged = data.merge(angle_df, left_on = "obs_tm", right_on = "obs_tm")
```

```python
fig, ax = plt.subplots(figsize=(12,10))
sns.regplot(x = "degree", y = "Forecast_error", data = data_merged)
print("Pearson-Correlation: \t {}\nSpearman_correlation: \t {}".format(stats.pearsonr(data_merged["degree"], data_merged["Forecast_error"]),
                                                                       stats.spearmanr(data_merged["degree"], data_merged["Forecast_error"])))
```
