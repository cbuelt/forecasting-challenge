```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date, datetime, timedelta
import random
from scipy import stats
```

```python
quantiles = [0.025, 0.25, 0.5, 0.75, 0.975]
horizons = [36, 48 ,60, 72, 84]
sns.set_theme(color_codes=True)
```

```python
data = pd.read_feather("data/berlin_data/historic_data/icon_eps_t_2m.feather")
#Filter for horizons
data = data[data["fcst_hour"].isin(horizons)]
data.dropna(inplace=True)
data.head()
```

# Visualize relationship between time and forecast error

```python
data["Forecast_error"] = data["obs"]-data["ens_mean"]
data
```

```python
fig, ax = plt.subplots(figsize=(12,10))
sns.lineplot(x = "obs_tm", y = "Forecast_error", hue = "fcst_hour", data=data)
```

```python
data["day"] = pd.DatetimeIndex(data["obs_tm"]).dayofyear
plot = data.groupby("day").mean().reset_index()
fig, ax = plt.subplots(figsize=(12,10))
sns.regplot(x = "day", y = "Forecast_error", data=plot)
print("Pearson-Correlation: \t {}\nSpearman_correlation: \t {}".format(stats.pearsonr(plot["day"], plot["Forecast_error"]),stats.spearmanr(plot["day"], plot["Forecast_error"])))
```

## Analysis
It is evident that there is some correlation between the forecast error and the time of the year. The correlation is small, but timely information could be implemented into the temperature model.
