```python
import pandas as pd
import numpy as np
from datetime import date, timedelta, datetime
```

# Aggregate data

```python
#Load predictions
current_date = date.today().strftime("%Y-%m-%d")
targets = ["DAX","temperature", "wind"]
```

```python
final_submission = pd.DataFrame(columns = ["forecast_date","target","horizon","q0.025","q0.25","q0.5","q0.75","q0.975"])
for target in targets:
    data = pd.read_pickle("single/{}_{}".format(target, current_date))
    final_submission = final_submission.append(data)
final_submission = final_submission.reset_index().drop("index", axis = 1)
```
